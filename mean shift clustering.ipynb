```python
import numpy as np 
import pandas as pd 
from sklearn.cluster import MeanShift 
from sklearn.datasets.samples_generator import make_blobs 
from matplotlib import pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
```

```python
df=pd.read_csv("seeds_dataset.csv")
df.head()
```

```python
     
clusters = [df['compactness'],df['perimeter'] ,df['lengthOfKernelGroove']] 
  
X, _ = make_blobs(n_samples = 100, centers = clusters, 
                                   cluster_std = 0.8) 
```

```python
ms = MeanShift() 
ms.fit(X) 
cluster_centers = ms.cluster_centers_ 
```

```python
fig = plt.figure() 
  
ax = fig.add_subplot(111, projection ='3d') 
  
ax.scatter(X[:, 0], X[:, 1], X[:, 2], marker ='o',color='green') 
  
ax.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
           cluster_centers[:, 2], marker ='x', color ='red', 
           s = 300, linewidth = 5, zorder = 10) 
  
plt.show() 
```
